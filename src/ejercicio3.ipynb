 ¿Cómo varía el tipo de gasto en comida según los ingresos de los clientes?
 Muestra visualmente que tipo de productos son los más consumidos por los
 clientes según su nivel de ingresos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(r"..\\data\\ml_project1_data.csv")
print(df.head())
```

```python
df_3=df.copy()
df_3=df_3.dropna(subset="Income")
df_3.isnull().sum()
```

```python
df_3= df_3 = df[["Income", "MntFishProducts", "MntFruits", "MntWines", "MntMeatProducts", "MntSweetProducts", "MntGoldProds"]]

print(df_3.head())
print(df_3.tail())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_3['MntFishProducts'], y=df_3['Income'], color='blue')

plt.title('Relación entre Ingresos y MntFishProducts')
plt.xlabel('MntFishProducts')
plt.ylabel('Income')


plt.show()
print(df_3[['Income', 'MntFishProducts']].corr())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_3['MntFruits'], y=df_3['Income'], color='blue')

plt.title('Relación entre Ingresos y MntFruits')
plt.xlabel('MntFruits')
plt.ylabel('Income')


plt.show()
print(df_3[['Income', 'MntFruits']].corr())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_3['MntWines'], y=df_3['Income'], color='blue')

plt.title('Relación entre Ingresos y MntWines')
plt.xlabel('MntWines')
plt.ylabel('Income')


plt.show()
print(df_3[['Income', 'MntWines']].corr())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_3['MntSweetProducts'], y=df_3['Income'], color='blue')

plt.title('Relación entre Ingresos y MntSweetProducts')
plt.xlabel('MntSweetProducts')
plt.ylabel('Income')


plt.show()
print(df_3[['Income', 'MntSweetProducts']].corr())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_3['MntGoldProds'], y=df_3['Income'], color='blue')

plt.title('Relación entre Ingresos y MntGoldProds')
plt.xlabel('MntGoldProds')
plt.ylabel('Income')


plt.show()
print(df_3[['Income', 'MntGoldProds']].corr())
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df_3[['Income'] + ["MntWines"]+ ["MntGoldProds"]+ ["MntSweetProducts"]+["MntMeatProducts"]+["MntFruits"]+["MntFishProducts"]].corr(), annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Matriz de Correlación entre Ingresos y Gasto en Alimentos')
plt.show()
```
