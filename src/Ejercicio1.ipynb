```python
import pandas as pd


df = pd.read_csv(r"..\\data\\ml_project1_data.csv")
print(df.head())
```

¿Existe una relación entre los ingresos anuales de los clientes y la cantidad
de dinero que gastan en iFood? Visualiza la relación con un gráfico y
describe tus conclusiones.


```python
df_estudiar=df.copy()
```

```python
df_estudiar=df_estudiar.dropna(subset="Income")
df_estudiar.isnull().sum()
```

```python
df_estudiar.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(8, 6))
sns.boxplot(x=df_estudiar['Income'], color='lightblue')
plt.title('Ingresos')
plt.show()
```

```python
#aqui lo que hago es sumar todos los gastos
df_estudiar['Total'] = df_estudiar['MntWines'] + df_estudiar['MntFruits'] + df_estudiar['MntMeatProducts'] + df_estudiar['MntFishProducts'] + df_estudiar['MntSweetProducts'] + df_estudiar['MntGoldProds']
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_estudiar['Total'], y=df_estudiar['Income'], color='blue')

plt.title('Relación entre Ingresos y Gasto Total en iFood')
plt.xlabel('Total')
plt.ylabel('Income')


plt.show()
```

```python
df_estudiar[['Income', 'Total']].corr()
```

Esta correlacion entre el total gastando en ifood y los ingresos es alta y positiva por lo que nos ha refleja que a mayor ingresos, mayor son los gastos en la tienda.

```python
df_estudiar.describe().T["75%"] - df_estudiar.describe().T["25%"]
```

Con esto comprobamos que la diferencia entre el cuartil del 75% y del 25% en algunos casos si que hay bastante diferencia por lo que tenemos que ver si los outliers son errores

```python
import pandas as pd

# Supongamos que tienes un DataFrame llamado df con una columna "Income"
Q1 = df_estudiar["Income"].quantile(0.25)  # Primer cuartil (25%)
Q3 = df_estudiar["Income"].quantile(0.75)  # Tercer cuartil (75%)
IQR = Q3 - Q1  # Rango intercuartílico

# Filtrar los datos eliminando los outliers
df_filtered = df_estudiar[(df_estudiar["Income"] >= Q1 - 1.5 * IQR) & (df_estudiar["Income"] <= Q3 + 1.5 * IQR)]
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_filtered['Income'], color='lightblue')
plt.title('Ingresos')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_filtered['Total'], y=df_filtered['Income'], color='blue')

plt.title('Relación entre Ingresos y Gasto Total en iFood')
plt.xlabel('Total')
plt.ylabel('Income')


plt.show()
```

```python
df_filtered[['Income', 'Total']].corr()
```

Una vez filtrados los datos y eliminado los nulos y filtrados los outliers, la correlacion es mucho mayor por lo que a mayor income, mayor gasto
